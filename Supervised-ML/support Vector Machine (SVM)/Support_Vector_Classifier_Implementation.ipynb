# Support Vector Classifier Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Dataset

```python
# Let's Create synthetic dataset
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=500,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0,random_state=2)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
```

```python
len(X[:,0]) , len(X[:,1])
```

### Split Data into Training and Test

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

## Training

```python
# Import SVM Library
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
# Prediction
y_predict = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print(classification_report(y_test,y_predict))
```

```python
print(confusion_matrix(y_test,y_predict))
```

### Try Different Kernels

```python
# rbf Kernel
rbf=SVC(kernel='rbf')
rbf.fit(X_train,y_train)
```

```python
y_predict_rbf=rbf.predict(X_test)
```

```python
# evaluation metrics
print(classification_report(y_test,y_predict_rbf))
```

```python
print(confusion_matrix(y_test,y_predict_rbf))
```

```python
# Polynomial Kernels
poly=SVC(kernel='poly')
poly.fit(X_train,y_train)
```

```python
y_pred_poly= poly.predict(X_test)
```

```python
# report Metrics
print(classification_report(y_test,y_pred_poly))
```

```python
print(confusion_matrix(y_test,y_pred_poly))
```

### HyperParameters

* `C` :
     1. High C  Low bias, high variance (tries to classify all training examples correctly, risking overfitting).
     2. Low C  High bias, low variance (allows some misclassification, leading to better generalization).

* `gamma` :
  1. High gamma  Model is more complex (each point has high influence, possible overfitting).
  2. Low gamma  Model is simpler (each point has less influence, possible underfitting).     


```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1,1,10,100,1000],
              'gamma': [1,0.1,0.01,0.001,0.0001],
              'kernel': ['rbf']
              }
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# predition
y_pred_grid = grid.predict(X_test)
# classification report
print(classification_report(y_test,y_pred_grid))
# confusion matrix
print(confusion_matrix(y_test,y_pred_grid))
```
