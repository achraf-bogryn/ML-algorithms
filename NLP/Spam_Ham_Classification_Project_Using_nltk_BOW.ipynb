# Bag Of Words(BOW)

apply BOW for spam messages

```python
import pandas as pd
messages = pd.read_csv('/content/SMSSpamCollection.tsv', sep='\t',
                           names=["label", "message"])
messages
```

## Data Cleaning

* Stop Words
* Limitization
* ...

```python
import numpy as np
import re  #Use For Regular expression
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
len(corpus)
```

## Create a Bag Of Words

```python
# create the Bag Of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500 , ngram_range=(1,2))
```

```python
# independent features
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000,
                    formatter=dict(float=lambda x: "%.3g" % x))
X[:10]
```

```python
cv.vocabulary_
```

```python
# Outpots Features
y = pd.get_dummies(messages['label'])
y
```

```python
y = y.iloc[:,1].values
```

## Train Text Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred = spam_detect_model.predict(X_train)
```

```python
from sklearn.metrics import recall_score , precision_score , confusion_matrix , classification_report
print(classification_report(y_train , y_pred))
```

```python
print("precision is : ",precision_score(y_train , y_pred))
print("recall is : ",recall_score(y_train , y_pred))
```
