# Bag Of Words(BOW)

apply BOW for spam messages

```python
import pandas as pd
messages = pd.read_csv('/content/SMSSpamCollection.tsv', sep='\t',
                           names=["label", "message"])
messages
```

## Data Cleaning

* Stop Words
* Limitization
* ...

```python
import numpy as np
import re  #Use For Regular expression
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
len(corpus)
```

## Create the Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
# For binary BOW binary=True
vectorizer = CountVectorizer(max_features=100,binary=True)
```

```python
X = vectorizer.fit_transform(corpus).toarray()
vectorizer.get_feature_names_out()
```

## N-Grams

```python
# by using N-grams parameter
vectorizer = CountVectorizer(max_features=500,binary=True,ngram_range=(2,3))
X = vectorizer.fit_transform(corpus).toarray()
# vectorizer.get_feature_names_out()
```
