### Computing Cost

The term 'cost' in this assignment might be a little confusing since the data is housing cost. Here, cost is a measure how well our model is predicting the target price of the house. The term 'price' is used for housing data.



```python
def compute_cost(x , y , w , b):
  m=x.shape[0];
  f_wb = np.zeros(m);
  cost_sum=0;
  for i in range(m):
    f_wb= w*x[i] + b;
    cost = (f_wb -y[i])**2;
    cost_sum += cost;
  total_cost = (1/(2*m))*cost_sum;
  return total_cost;
```

## Gradient Descent for Linear Regression

```python
def compute_gradient(x, y , w,b):
  m = x.shape[0];
  f_wb=np.zeros(m);
  dj_dw = 0;
  dj_db = 0;
  for i in range(m):
    f_wb = w*x[i] + b;
    dj_dw_i = (f_wb - y[i])*x[i];
    dj_db_i = f_wb - y[i];
    dj_dw += dj_dw_i;
    dj_db += dj_db_i;
  dj_dw = dj_dw/m;
  dj_db = dj_db/m;
  return dj_dw, dj_db;
```

```python
import copy # Import the copy module
import math
def gradient_descendent(x,y , w_in , b_in , alpha , num_iteration,compute_cost , compute_gradient):
  w = copy.deepcopy(w_in) # avoid modifying global w_in
  w=w_in;
  b=b_in;
  J_history = []
  p_history = []
  for i in range(num_iteration):
    dj_dw, dj_db = compute_gradient(x,y,w,b);
    w = w - alpha*dj_dw;
    b = b - alpha*dj_db;

    # cost = np.mean((w*x + b - y)**2);
    # J_history.append(cost);

    if i<10000:
      J_history.append(compute_cost(x,y,w,b));
      p_history.append([w,b])

      # Print cost every at intervals 10 times or as many iterations if < 10
    if i% math.ceil(num_iteration/10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")


    # if i % 100 == 0:
    #         print(f"Iteration {i}: Cost = {cost}, w = {w}, b = {b}")

  return w, b, J_history , p_history

```

```python
# initialize parameters
w_init = 0
b_init = 0
# some gradient descent settings
iterations = 10000
tmp_alpha = 1.0e-2
# run gradient descent
w_final, b_final, J_hist, p_hist = gradient_descendent(x_train1 ,y_train1, w_init, b_init, tmp_alpha,
                                                    iterations, compute_cost, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

## Linear Regression from Scratch

```python
import numpy as np

class LinearRegression:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        """
        Initialize the Linear Regression model.

        Parameters:
        - learning_rate: float, the step size for gradient descent (default: 0.01)
        - n_iterations: int, the number of iterations for gradient descent (default: 1000)
        """
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        """
        Train the Linear Regression model using Gradient Descent.

        Parameters:
        - X: numpy array, input features (shape: [n_samples, n_features])
        - y: numpy array, target values (shape: [n_samples])
        """
        n_samples, n_features = X.shape

        # Initialize weights and bias to zeros
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient Descent
        for _ in range(self.n_iterations):
            # Predictions
            y_pred = np.dot(X, self.weights) + self.bias

            # Compute gradients
            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))  # Gradient for weights
            db = (1 / n_samples) * np.sum(y_pred - y)        # Gradient for bias

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        """
        Predict the target values for the given input features.

        Parameters:
        - X: numpy array, input features (shape: [n_samples, n_features])

        Returns:
        - y_pred: numpy array, predicted target values (shape: [n_samples])
        """
        return np.dot(X, self.weights) + self.bias

# Example usage
if __name__ == "__main__":
    # Generate some sample data
    np.random.seed(0)
    X = 2 * np.random.rand(100, 1)
    y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3X + noise

    # Flatten y to 1D array
    y = y.flatten()

    # Create and train the model
    model = LinearRegression(learning_rate=0.1, n_iterations=1000)
    model.fit(X, y)

    # Make predictions
    X_new = np.array([[0], [2]])
    y_pred = model.predict(X_new)

    print("Predictions:", y_pred)
    print("Weights:", model.weights)
    print("Bias:", model.bias)
```
