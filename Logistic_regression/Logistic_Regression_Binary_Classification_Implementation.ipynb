# Logistic Regression Binary Classificatio Problem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Export Dataset

```python
from sklearn.datasets import make_classification
```

```python
# create a dataset
X,y = make_classification(n_samples=1000 , n_features=10 , n_classes=2 , random_state=42)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

## Training Our Model

```python
# Model Training
from sklearn.linear_model import LogisticRegression
logistic  = LogisticRegression()
logistic.fit(X_train,y_train)
```

```python
# do some prediction
y_pred = logistic.predict(X_test)
y_pred
```

```python
logistic.predict_proba(X_test)[:10]
```

```python
# n=we ave 300 line and each line have 2 column cnain probability of each class
logistic.predict_proba(X_test).shape
```

```python
# Evaluate Our Model
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report , recall_score , precision_score
score = accuracy_score(y_test,y_pred)
print(score)
```

```python
CM = confusion_matrix(y_test,y_pred)
CM
```

```python
print(classification_report(y_test,y_pred))
```

## HyperParameter Tuning And Cross Validation

```python
model = LogisticRegression()
penalty = ['l1','l2', 'elasticnet']
c_values = [100 , 20 ,1.0 , 0.1 , 0.001]
solver = ['newton-cg', 'lbfgs', 'liblinear','sag','saga']
```

```python
params = dict(penalty= penalty , C=c_values , solver=solver)
```

```python
from sklearn.model_selection import  StratifiedKFold
cv= StratifiedKFold()
```

```python
# GridSearchCV
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=model , param_grid=params,scoring='accuracy' , cv=cv , n_jobs=-1)
grid_search
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
y_predict = grid_search.predict(X_test)
scores = accuracy_score(y_test,y_predict)
print(score)
print(classification_report(y_test,y_predict))
```

```python
cm = confusion_matrix(y_test,y_predict)
cm
```

### Optimize Our model By add best parametrs


```python
model_optimize = LogisticRegression(penalty='l1',solver='liblinear',C=0.1)
model_optimize.fit(X_train,y_train)
```

```python
y_pred_op = model_optimize.predict(X_test)
score_op = accuracy_score(y_test,y_pred_op)
print(score_op)
print(classification_report(y_test,y_pred_op))
```

# Logistic Regression for Imbalanced Dataset

```python
from collections import Counter
from sklearn.datasets import make_classification

# imblanced Dataset
X,y = make_classification(n_samples=10000 ,
                          n_features=2 , n_clusters_per_class=1,
                          n_redundant=0 , weights=[0.99],random_state=42)
```

```python
pd.DataFrame(y).value_counts()
```

```python
# 1% = 100 ---> 9% = 900
# 10% = 1000 ---> 90% = 9000
# 99
```

```python
sns.scatterplot(x= pd.DataFrame(X)[0],y= pd.DataFrame(X)[1],hue=y)
```

Like we can say in this visualization , oue data that it's not blanced betweeen this two category

* `How we can train model with this data ?`

```python
# 1 - Split data into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# 2 - start training and use different HyperParameters Tuning
model = LogisticRegression()
penalty = ['l1','l2', 'elasticnet']
c_values = [100 , 20 ,1.0 , 0.1 , 0.001]
solver = ['newton-cg', 'lbfgs', 'liblinear','sag','saga']
# here like loop inside loop for each w wait y loop finish
class_weight = [{0:w , 1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
parameters = dict(penalty= penalty , C=c_values , solver=solver , class_weight=class_weight)
```

```python
parameters
```

```python
# shose Cross Validation way
cv = StratifiedKFold()
# GridSearchCV to find best params
gridSearchModel = GridSearchCV(estimator=model , param_grid=parameters ,scoring='accuracy' , cv=cv , n_jobs=-1)
gridSearchModel.fit(X_train,y_train)
```

```python
gridSearchModel.best_params_
```

```python
print("------------------scores of binary classification with imblanced dataset----------------------")
y_pred= gridSearchModel.predict(X_test)
acuracy_score = accuracy_score(y_test,y_pred)
print("accuracy score: ",acuracy_score)
print("-------------------------------------------------------------------")
print("classification report :  ")
print(classification_report(y_test,y_pred))
print("--------------------------------------------------------------------")
cm=confusion_matrix(y_test,y_pred)
print("confusion matrix :  ")
print(cm)
print("--------------------------------------------------------------------")
```

# Logistic Regression With ROC curve and ROC AUC score

```python
#  import roc curve and auc
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot
```

```python
# generate anew dataset with 2 classes
X,y=make_classification(n_samples=1000 , n_classes=2 , random_state=42)
```

```python
# split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# create a dummy model with defualt 0 as output (majority class) -> we can say initial value on test data afer we pull by probability
dummy_model_prob =[0 for _ in range(len(y_test))]
dummy_model_prob[:10]
```

```python
# Let's create a basic Logistic Regression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
# predcition based on probabilities
model_prob = model.predict_proba(X_test)
```

```python
# Let's focus on positif outcome
model_prob = model_prob[:,1]
```

```python
# Let's calculate score
dummy_model_auc = roc_auc_score(y_test,dummy_model_prob)
model_auc = roc_auc_score(y_test,model_prob)
print("dummy model auc: ",dummy_model_auc)
print("model auc: ",model_auc)
```

```python
# calcule ROC curve
dummy_model_fpr , dummy_model_tpr , _ = roc_curve(y_test,dummy_model_prob)
model_fpr , model_tpr , thresholds = roc_curve(y_test,model_prob)
thresholds
```

roc curve he give us `False positive rate(FPR)` and `true positive rate(TPR)`that why we use him.

```python
# Plot the roc curve for the model
pyplot.plot(dummy_model_fpr,dummy_model_tpr,linestyle='--',label='dummy model')
pyplot.plot(model_fpr,model_tpr,marker='.',label='Logistic')

# axis label
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')

# show the legend
pyplot.legend()

# show the plot
pyplot.show()
```

How we choose the threshold Base on this curce ?

* we need to best threshold from this curce that he will give best result .

```python
# Plot ROC curves
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.plot(dummy_model_fpr, dummy_model_tpr, linestyle='--', label='Dummy Model (AUC = %0.2f)' % dummy_model_auc)
plt.plot(model_fpr, model_tpr, marker='.', label='Logistic Regression (AUC = %0.2f)' % model_auc)

# Annotate thresholds on the ROC curve
for i, threshold in enumerate(thresholds):
    if i % 5 == 0:  # Adjust annotation frequency if needed
        plt.text(model_fpr[i], model_tpr[i], np.round(threshold, 2), fontsize=8)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
